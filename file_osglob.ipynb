ディレクトリ（フォルダ）内のファイルの一覧表示にはこんな方法もある

*   ワイルドカードによりパターンマッチング
*   正規表現

```python
import glob

files = glob.glob('*')
for file in files:
  print(file)
```

ファイルのみ，ディレクトリのみの一覧を取得したい場合：よくある．．．

path.isfile関数を使用してファイルのみ取得

path.isdir関数　ディレクトリのみ

```python
import os

filelist=[]

#files = glob.glob('./drive/MyDrive/*')
for file in os.listdir('./drive/MyDrive'):
  if os.path.isfile(os.path.join('./drive/MyDrive',file)):
    filelist.append(file)

print(filelist)
```

```python
import os

filelist=[]

for file in os.listdir('./drive/MyDrive'):
  if os.path.isdir(os.path.join('./drive/MyDrive',file)):
    filelist.append(file)

print(filelist)
```

```python
import os

#パスが長くて面倒なので
path = './drive/MyDrive'
filelist=[]

for file in os.listdir(path):
  if os.path.isdir(os.path.join(path,file)): #os.path.-, os.path.j- のpathは上記パスではありません
    filelist.append(file)

print(filelist)
```
